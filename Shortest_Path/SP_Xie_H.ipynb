```python
import numpy as np
import math
```

```python
def shortest_path(file_name):
    # initialize
    s = eval(input("Enter the start vertex:"))
    adjacent = np.loadtxt(file_name, delimiter=",")
    size = adjacent.shape[0]
    distTo = [math.inf]*size
    edgeTo = [None]*size
    distTo[s] = 0
    L = []
    L.append(s)

    def find_min_dist():
        index = -1
        min_dist = math.inf
        for i in L:
            if distTo[i] < min_dist:
                min_dist = distTo[i]
                index = i
        return index

    def update_list():
        while L:
            u = find_min_dist()
            for i in range(size):
                if adjacent[u][i] != 0 and i not in L:
                    L.append(i)
            L.remove(u)
            for i in L:
                if adjacent[u][i] != 0 and distTo[u] + adjacent[u][i] < distTo[i]:
                    distTo[i] = distTo[u] + adjacent[u][i]
                    edgeTo[i] = u
        print("distTo list: ",distTo)
        print("edgeTo list: ",edgeTo)
        print()
        
        def output():
            vertex = []
            for i in range(size):
                if i!=s:
                    if edgeTo[i] != None:
                        print("shortest path from {} to {}:".format(s,i), end = " ")
                        j = edgeTo[i]
                        str1 = str(j)+" -> "+str(i)
                        while edgeTo[j] != None:
                            j = edgeTo[j]
                            str1 = str(j)+" -> "+str1
                        print(str1)
                        print("length: {}".format(distTo[i]))
                        print()

        return output()

    return update_list()
```
